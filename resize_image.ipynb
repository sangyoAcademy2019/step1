```
from PIL import Image
import math
import os
```

```
# 
tmp_dir = '/content/drive/My Drive/Colab Notebooks/temp'
re_dir  = '/content/drive/My Drive/Colab Notebooks/img'
re_width = 360
re_height = 240
```

```
# 画像を小さくする
for file in os.listdir(tmp_dir):
    try:
        img = Image.open(tmp_dir + '/' + file)
        width = img.size[0]
        height = img.size[1]
        if width < height:
            img = img.transpose(Image.ROTATE_90)
        re_img = img.resize((re_width, re_height))
        re_img.save(re_dir+'/'+file)
        img = None
        re_img = None
    except Exception as e:
        print('exception*{}'.format(e))    
```
